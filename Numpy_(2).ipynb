```python
# 8) Розбити масив по вертикалі на 2 рівні частини (не використовувати абсолютні числа, мають бути два массиви по 4 колонки)
```

```python
a_left =a_without_target[:, :2]  # Перші 2 колонки
a_right = a_without_target[:, 2:4] # Наступні 2 колонки
```

```python
print('Перші 2 колонки\n', a_left)
print('Наступні 2 колонки\n', a_right)
```

```python
# 10) Відсортувати обидва массиви по 1-й колонці: 1-й за збільшенням, 2-й за зменшенням
```

```python
a_left_sorted = a_left[a_left[:, 0].argsort()]
a_right_sorted = a_right[a_right[:, 0].argsort()[::-1]]
```

```python
print('отсортированный первый массив по возрастанию, первая колонка\n', a_left_sorted )
print('отсортированный второй массив по убыванию, первая колонка\n', a_right_sorted )
```

```python
# 11) Зібрати обидва массиви в одне ціле
```

```python
a_combined = np.hstack((a_left_sorted, a_right_sorted))
```

```python
print('Объединенные массивы', a_combined)
```

```python
# 12) Знайти найбільш часто повторюване значення в массиві
```

```python
uniqs, counts = np.unique(a_combined, return_counts=True)
most_frequent_index = np.argmax(counts)
most_frequent_value = uniqs[most_frequent_index]
print('найбільш часто повторюване значення в массиві : ', most_frequent_value)
```

```python
# 13) Написати функцію, яка б множила всі значення в колонці, які менше середнього значения в цій колонці, на 2, і ділила інші значення на 4.
```

```python
def modify_column(array):
    mean_value = np.mean(array)
    modified_array = np.where(array < mean_value, array * 2, array / 4)
    return modified_array

```

```python
# Застосувати отриману функцію до 3-ї колонки
```

```python
modified_column = modify_column(a_combined[:,2])
print("Модифікована колонка:", modified_column)
```
